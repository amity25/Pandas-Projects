```python
# Create Your First DataFrame
# Import pandas
import pandas as pd

# Creating dictionary data
data = {
    "name": ["Amit", "Rahul", "Priya"],
    "sales": [1000, 2000, 1500]
}

# Convert dictionary into DataFrame
df = pd.DataFrame(data)

# Print DataFrame
print(df)
```

```python
print(df["sales"])
```

```python
# Show all column names
print(df.columns)
```

```python
# Get name and sales columns together
print(df[["name", "sales"]])
```

```python
# iloc means integer location
# iloc[0] = first row

print(df.iloc[0])
```
